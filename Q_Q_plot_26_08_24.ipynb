### A Q-Q (quantile-quantile) plot is a graphical tool used to compare two probability distributions by plotting their quantiles against each other. It is frequently employed to determine if a dataset adheres to a particular distribution, like the normal distribution.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Generate a random matrix
data = np.random.normal(loc=0, scale=1, size=(100, 100))

# Flatten the matrix to get all elements in one array
data_flattened = data.flatten()

# Q-Q plot
stats.probplot(data_flattened, dist="norm", plot=plt)
plt.show()

# Shapiro-Wilk test
shapiro_test = stats.shapiro(data_flattened)
print(f"Shapiro-Wilk Test p-value: {shapiro_test.pvalue}")
```

```python
# Generate 5*5 random matrix and flattened it into an 1D-array
A = np.random.rand(5,5)
A_vec = A.flatten()
A_vec
```

### probplot - function in scipy.stats module, which calculates quantiles for a probability plot and optionally shows the plot. Link for its documentation - https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.probplot.html
