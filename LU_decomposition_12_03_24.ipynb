### Code to perform LU-decomposition
#### Here I have implimented the LU-faactorisation from Section 2.3 of "An Introduction to Numerical Analysis" by Endre Suli and David F. Mayers

```python
# import all the libraries
import numpy as np
```

```python
# Given matrix A
A = np.array([[1, -2, 4],
              [-5, 2, 0],
              [1, 0, 3]])
n = A.shape[0]
```

```python
# Initalize the lower and upper traiangular matrices
L = np.eye(n)
U = np.zeros_like(A)
```

```python
U[0, :] = A[0, :]   # As U(1, j) = A(1, j)
L[1:, 0] = (1 / U[0, 0]) * A[1:, 0]  # L(1, j) = A(1, j) / U(1, 1)

# Now we calculate U(i,j) and L(i,j) by equation 2.18 and 2.19
for i in range(1, n):
    for j in range(i):
        L[i, j] = (1 / U[j, j]) * (A[i, j] - np.dot(L[i, :j], U[:j, j]))
    for j in range(i, n):
        U[i, j] = A[i, j] - np.dot(L[i, :i], U[:i, j])
```

```python
print('Lower triangular matrix (L):')
print(L)
print('Upper triangular matrix (U):')
print(U)
```
